```python
import json
```

```python
def data_load(filename):
    with open(filename , "r") as f:
        data = json.load(f)

    return data
```

```python
data = data_load("store_data.json")
print(data)
print(type(data))
```

```python
def clean_data(data):
    text_to_num = {"one" : 1, "two" : 2, "three" : 3, "four" : 4, "five" : 5}
    cleaned_data = []
    unique_users = set()

    for user in data:
        # Clean ratings - data consistency
        raw_rating = str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        # Handle missing vals
        raw_age = user.get("age")
        if raw_age == None:
            user["age"] = None

        # Deduplication
        if user["name"].strip() in unique_users:
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
        
```

```python
data = clean_data(data)
```

```python
# Get meaningful insights from data
def get_insights(data):
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])

    print(f"avg rating = {tot_rating / len(data)}")

    # percantage of user with poor rating

    poor_rating = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating += 1

    print(f"% of user with poor rating = {poor_rating/len(data) * 100 }")
```

```python
get_insights(data)
```

```python
# Build recommendation Feature
def get_recommendtions(data):
    recommendations = []

    for user in data:
        curr_recommm = {}
        curr_recomm["name"] = user["name"]

        if (float(user["rating"]) >= 4):
            curr_recomm["Brand"] = "Apple"
        else:
            curr_recomm["Brand"] = "Samsung"

        recommendations.append(curr_recomm)
    return recommendations

```
